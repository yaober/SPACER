```python
import scanpy as sc
import pandas as pd
import numpy as np
```

```python

adata = sc.read_h5ad('/project/DPDS/Wang_lab/shared/spatial_TCR/data/train_validate/Spatial_TCR_Visium/p16/p16.h5ad')
tcr_count = pd.read_csv('/project/DPDS/Wang_lab/shared/spatial_TCR/data/train_validate/Spatial_TCR_Visium/p16/TCR_count_table_p16.csv')
```

```python
tcr_count
```

```python
adata.obs
```

```python
tcr_count.drop(columns=['clonotypes','count'], inplace=True)
tcr_count
```

```python
tcr_count.drop_duplicates(inplace=True)
```

```python
tcr_count
```

```python
tcr_count['T'] = 1
```

```python
tcr_count.set_index('barcode', inplace=True)
```

```python
adata.obs.index = adata.obs.index.str.split('-').str[0]
```

```python
adata.obs
```

```python
overlap_indices = adata.obs_names.intersection(tcr_count.index)
overlap_indices
```

```python
adata.obs['T'] = 0
```

```python
adata.obs.loc[overlap_indices, 'T'] = 1
```

```python
adata.obs[adata.obs['T']== 1]
```

```python
adata.obs['X'] = adata.obs['x_array'] *100
adata.obs['Y'] = adata.obs['y_array'] *100
adata.obs
```

```python
adata.obs['cell_type'] = adata.obs['T']+1
```
