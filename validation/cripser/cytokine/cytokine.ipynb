```python
import pandas as pd
```

```python
results = pd.read_csv('validation/cripser/cytokine/science.abj4008_table_s2.csv')
results
```

```python
results.drop(columns=['Unnamed: 10','Unnamed: 11','Unnamed: 12'], inplace=True)
results
```

```python
results['Hit'].value_counts()
```

```python
results = results[results['Hit']==True]
results
```

```python
results.columns.to_list
```

```python
grouped = results.groupby('Gene')
```

```python
filtered = grouped.filter(lambda x: x['Hit_Type'].nunique() == 1).drop_duplicates(subset='Gene', keep='first')
```

```python
filtered
```

```python
final_results = filtered[['Gene', 'Hit_Type']]
```

```python
final_results['Gene'].nunique()
```

```python
final_results
```

```python
diff_gene = pd.read_csv('validation/diff_gene/all_data_diff_gene_markers_top5000_filtered.csv')
diff_gene
```

```python
final_results = final_results.merge(diff_gene, left_on='Gene', right_on='gene', how='left')
```

```python
final_results
```

```python
final_results.dropna(inplace=True)
final_results
```

```python
final_results['Hit_Type'] = final_results['Hit_Type'].apply(lambda x: 1 if x == 'Negative Hit' else -1)
```

```python
final_results
```

```python
final_results['cosistancy'] = final_results['Hit_Type'] * final_results['logfoldchange']
```

```python
final_results
```

```python
final_results['cosistancy'] = final_results['cosistancy'].apply(lambda x: True if x > 0 else False)
```

```python
final_results
```

```python
final_results['cosistancy'].value_counts()
```

```python
final_results.to_csv('validation/cripser/cytokine/cripser_cytokine_final_results.csv', index=False)
```
