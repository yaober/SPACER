```python
import pandas as pd
```

```python
df = pd.read_csv('validation/cripser/key_regulators/mageck.csv')
df
```

```python
df = df[(df['neg|p-value'] < 0.05) | (df['pos|p-value'] < 0.05)]
df = df[(df['neg|fdr'] < 0.05) | (df['pos|fdr'] < 0.05)]
df
```

```python
diff_gene = pd.read_csv('validation/diff_gene/all_data_diff_gene_markers_top5000_filtered.csv')
diff_gene
```

```python
df = df.merge(diff_gene, left_on='id', right_on='gene', how='left')
df
```

```python
df.dropna(subset=['gene'], inplace=True)
df
```

```python
df['consistancy'] = df.apply(lambda x: True if x['neg|lfc'] * x['logfoldchange'] > 0 else False, axis=1)
```

```python
df
```

```python
df['consistancy'].value_counts()
```

```python
df.to_csv('validation/cripser/key_regulators/final_results_key_regulators.csv', index=False)
```
