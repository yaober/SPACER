```python

# Define the MAGeCK result files and their corresponding conditions
files_conditions = {
    "validation/cripser/tcell_migration/MvsB.csv": "MvsB",  # Meninges vs Blood
    "validation/cripser/tcell_migration/PvsB.csv": "PvsB",  # Parenchyma vs Blood
    "validation/cripser/tcell_migration/MvsS.csv": "MvsS",  # Meninges vs Spleen
    "validation/cripser/tcell_migration/PvsS.csv": "PvsS",  # Parenchyma vs Spleen
    "validation/cripser/tcell_migration/PvsM.csv": "PvsM",  # Parenchyma vs Meninges
    "validation/cripser/tcell_migration/SvsB.csv": "SvsB"   # Spleen vs Blood
}
```

```python

# Read and aggregate MAGeCK results from all files
mageck_results = []
for file, condition in files_conditions.items():
    df = pd.read_csv(file)  # Load each CSV file
    df["Condition"] = condition  # Add a new column indicating the condition
    mageck_results.append(df)

# Combine all MAGeCK results into a single DataFrame
mageck_combined = pd.concat(mageck_results, ignore_index=True)
mageck_combined
```

```python
mageck_combined['id'] = mageck_combined['id'].str.upper()
mageck_combined
```

```python
mageck_combined = mageck_combined[(mageck_combined['neg|p-value'] < 0.05) | (mageck_combined['pos|p-value'] < 0.05)]
mageck_combined
```

```python
mageck_combined.dropna(subset=['Figure'],inplace=True)
```

```python
mageck_combined
```

```python
aggregation_rules = {
    "neg|score": "mean",           
    "neg|p-value": "mean",       
    "neg|fdr": "mean",             
    "neg|goodsgrna": "mean",       
    "neg|lfc": "mean",            
    "pos|score": "mean",          
    "pos|p-value": "mean",
    "pos|fdr": "mean",
    "pos|goodsgrna": "mean",
    "pos|lfc": "mean",
    "Condition": lambda x: ";".join(set(x)),  
}

merged_mageck = mageck_combined.groupby("id").agg(aggregation_rules).reset_index()
```

```python

# Load your gene table
gene_table = pd.read_csv("validation/diff_gene/all_data_diff_gene_markers_top5000_filtered.csv")

# Merge the gene table with the combined MAGeCK results based on the gene name
merged = pd.merge(gene_table, merged_mageck, left_on="gene", right_on="id", how="left")

# Validate if logfoldchange (LFC) values are consistent between your table and MAGeCK results
merged["LFC_consistent"] = (merged["logfoldchange"] * merged["neg|lfc"] > 0)

```

```python
merged
```

```python
merged.dropna(subset=['id'], inplace=True)
```

```python
import matplotlib.pyplot as plt

```

```python
merged = merged.sort_values(by="logfoldchange")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(merged["logfoldchange"], merged["neg|lfc"], label="neg|lfc vs logfoldchange")

# Add labels, title, and legend
plt.xlabel("logfoldchange", fontsize=12)
plt.ylabel("neg|lfc", fontsize=12)
plt.title("Log-Fold Change vs Negative LFC", fontsize=14)
plt.axhline(0, color='grey', linestyle='--', linewidth=0.8)  # Add a horizontal line at y=0 for reference
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(merged["logfoldchange"], merged["neg|lfc"], alpha=0.7, edgecolor='k', s=50)

# Add labels, title, and trend line
plt.xlabel("logfoldchange", fontsize=12)
plt.ylabel("neg|lfc", fontsize=12)
plt.title("Scatter Plot of Log-Fold Change vs Negative LFC", fontsize=14)
```

```python
#calculate the coefficient between the logfoldchange abd neg|lf

import numpy as np

# Calculate the correlation coefficient between logfoldchange and neg|lfc

correlation = np.corrcoef(merged["logfoldchange"], merged["neg|lfc"])[0, 1] 
correlation
```
