```python
import pandas as pd
```

```python
genomewide_results = pd.read_csv('validation/cripser/T_cell_exhaustion/genomewide_results.csv')
```

```python
# Adjust the thresholds to be less strict for a broader selection
fdr_threshold = 0.1  # Allow slightly higher FDR
lfc_threshold = 0.3  # Lower the threshold for Log Fold Change
z_score_threshold = 2  # Include genes with slightly lower Z-scores

# Reapply the filtering with the relaxed thresholds
relaxed_filtered_genes = genomewide_results[
    (
        (genomewide_results['B16_tumor_merge_fdr'] < fdr_threshold) &
        (genomewide_results['B16_tumor_merge_lfc_diff'].abs() > lfc_threshold) &
        (genomewide_results['B16_tumor_merge_z'].abs() > z_score_threshold)
    ) |
    (
        (genomewide_results['MC38_T_v_I_merge_fdr'] < fdr_threshold) &
        (genomewide_results['MC38_T_v_I_merge_lfc_diff'].abs() > lfc_threshold) &
        (genomewide_results['MC38_T_v_I_merge_z'].abs() > z_score_threshold)
    )
]


# Display the first few rows of the relaxed filtering results
relaxed_filtered_genes
```

```python
relaxed_filtered_genes.to_csv('validation/cripser/T_cell_exhaustion/relaxed_filtered_genes.csv', index=False)
```

```python
strict_filtered_genes = genomewide_results[
    (
        # Significant in B16 melanoma model
        (genomewide_results['B16_tumor_merge_fdr'] < 0.05) &
        (genomewide_results['B16_tumor_merge_lfc_diff'].abs() > 0.5) &
        (genomewide_results['B16_tumor_merge_z'].abs() > 3)
    ) &
    (
        # Significant in MC38 colorectal cancer model
        (genomewide_results['MC38_T_v_I_merge_fdr'] < 0.05) &
        (genomewide_results['MC38_T_v_I_merge_lfc_diff'].abs() > 0.5) &
        (genomewide_results['MC38_T_v_I_merge_z'].abs() > 3)
    ) &
    (
        # Significant in in vitro experiments
        (genomewide_results['invitro_c_v_i_merge_fdr'] < 0.05) &
        (genomewide_results['invitro_c_v_i_merge_lfc_diff'].abs() > 0.5) &
        (genomewide_results['invitro_c_v_i_merge_z'].abs() > 3)
    )
]


# Display the filtered genes
strict_filtered_genes
```

```python
strict_filtered_genes.to_csv('validation/cripser/T_cell_exhaustion/strict_filtered_genes.csv', index=False)
```

```python
diffex_genes = pd.read_csv('validation/diff_gene/all_data_diff_gene_markers_top5000_filtered.csv')
```

```python
strict_filtered_genes = strict_filtered_genes.merge(diffex_genes[['gene', 'logfoldchange']], left_on='Gene',
                                                    right_on='gene',
                                                    how='left')
```

```python
strict_filtered_genes.dropna(subset=['logfoldchange'], inplace=True)
```
